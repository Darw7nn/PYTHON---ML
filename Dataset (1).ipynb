```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px

try:
    df = pd.read_csv('/content/Teen_Mental_Health_Dataset.csv')

    # Processamento de métricas
    df['mental_health_index'] = (df['stress_level'] + df['anxiety_level'] + (df['depression_label'] * 5)) / 3
    df['sleep_group'] = pd.cut(df['sleep_hours'], bins=[0, 5, 7, 12], labels=['Crítico (<5h)', 'Moderado (5-7h)', 'Ideal (>7h)'])

    # --- VISUALIZAÇÃO 1: TABELA ESTRATÉGICA ---
    metrics_summary = {
        'KPI ESTRATÉGICO': ['Média de Idade', 'Tempo de Sono (Médio)', 'Uso de Mídia Social', 'Nível de Estresse', 'Nível de Ansiedade', 'Índice de Risco Geral'],
        'VALOR ATUAL': [
            f"{df['age'].mean():.1f} anos",
            f"{df['sleep_hours'].mean():.1f}h",
            f"{df['daily_social_media_hours'].mean():.1f}h",
            f"{df['stress_level'].mean():.2f}/10",
            f"{df['anxiety_level'].mean():.2f}/10",
            f"{df['mental_health_index'].mean():.2f}/10"
        ]
    }
    fig_table = go.Figure(data=[go.Table(
        header=dict(values=['<b>KPI ESTRATÉGICO</b>', '<b>MÉTRICA APURADA</b>'], fill_color='#00CCFF', align='left', font=dict(color='black', size=15)),
        cells=dict(values=[metrics_summary['KPI ESTRATÉGICO'], metrics_summary['VALOR ATUAL']], fill_color='#1A1A1A', align='left', font=dict(color='cyan', size=13), height=30))
    ])
    fig_table.update_layout(title='RELATÓRIO EXECUTIVO: SAÚDE MENTAL GERAÇÃO Z', template='plotly_dark')
    fig_table.show()

    # --- VISUALIZAÇÃO 2: GRÁFICO DE PIZZA (DONUT) ---
    sleep_dist = df['sleep_group'].value_counts().reset_index()
    sleep_dist.columns = ['Grupo', 'Contagem']
    fig_pie = go.Figure(data=[go.Pie(labels=sleep_dist['Grupo'], values=sleep_dist['Contagem'], hole=.4,
                                    marker=dict(colors=['#00FF99', '#FFFF66', '#FF3399'], line=dict(color='#000000', width=2)))])
    fig_pie.update_layout(title='DISTRIBUIÇÃO DA QUALIDADE DO SONO', template='plotly_dark')
    fig_pie.show()

    # --- NOVA TABELA: RISCO POR PLATAFORMA ---
    platform_risk = df.groupby('platform_usage')['mental_health_index'].mean().reset_index().sort_values(by='mental_health_index', ascending=False)
    fig_platform = go.Figure(data=[go.Table(
        header=dict(values=['<b>PLATAFORMA</b>', '<b>ÍNDICE DE RISCO MÉDIO</b>'], fill_color='#FF3399', align='left', font=dict(color='white', size=15)),
        cells=dict(values=[platform_risk['platform_usage'], platform_risk['mental_health_index'].round(2)], fill_color='#1A1A1A', align='left', font=dict(color='#FFCCFF', size=13), height=30))
    ])
    fig_platform.update_layout(title='ANÁLISE DE RISCO POR PLATAFORMA', template='plotly_dark')
    fig_platform.show()

    # --- NOVA TABELA: IMPACTO DO SONO ---
    sleep_impact = df.groupby('sleep_group')[['stress_level', 'anxiety_level']].mean().reset_index()
    fig_sleep_table = go.Figure(data=[go.Table(
        header=dict(values=['<b>GRUPO DE SONO</b>', '<b>ESTRESSE</b>', '<b>ANSIEDADE</b>'], fill_color='#00FF99', align='left', font=dict(color='black', size=15)),
        cells=dict(values=[sleep_impact['sleep_group'], sleep_impact['stress_level'].round(2), sleep_impact['anxiety_level'].round(2)], fill_color='#1A1A1A', align='left', font=dict(color='#CCFFCC', size=13), height=30))
    ])
    fig_sleep_table.update_layout(title='MATRIZ DE IMPACTO DO SONO', template='plotly_dark')
    fig_sleep_table.show()

    # --- VISUALIZAÇÃO: GRÁFICO DE DISPERSÃO ---
    fig_scatter = px.scatter(df, x='daily_social_media_hours', y='mental_health_index', color='platform_usage',
                             trendline='ols', title='CORRELAÇÃO: TEMPO DE TELA VS. RISCO MENTAL',
                             labels={'daily_social_media_hours': 'Horas de Mídia Social', 'mental_health_index': 'Índice de Risco'},
                             template='plotly_dark')
    fig_scatter.show()

    print("✅ Dashboard atualizado e traduzido com sucesso!")
except Exception as e:
    print(f"❌ Erro: {e}")
```

# 📑 Conclusão Estratégica Atualizada

### **Insights Adicionais:**
1. **Plataformas Críticas:** Algumas plataformas apresentam índices de risco superiores à média, sugerindo que o tipo de conteúdo consome influência mais do que o tempo.
2. **Privação de Sono:** Adolescentes com menos de 5 horas de sono apresentam picos significativos de estresse e ansiedade, validando o sono como pilar fundamental da saúde mental.
3. **Recomendação:** Intervenções de bem-estar devem focar em 'higiene do sono' e educação sobre o uso consciente de plataformas específicas.
