## 1️⃣ Operações matemáticas

```python

a = 15
b = 4

print("Soma:", a + b)
print("Subtração:", a - b)
print("Multiplicação:", a * b)
print("Divisão:", a / b)
print("Potência:", a ** b)
```

## 2️⃣ Gerando números aleatórios

```python

import random

numeros = [random.randint(1,100) for _ in range(10)]
print("Lista aleatória:", numeros)
```

## 📈 3️⃣ Gráfico de linha

```python

import matplotlib.pyplot as plt

valores = [2,4,6,8,10,7,5]

plt.figure()
plt.plot(valores)

plt.title("Exemplo de Gráfico de Linha")
plt.xlabel("Tempo")
plt.ylabel("Valor")

plt.show()
```

## 📊 4️⃣ Gráfico de barras

```python

import matplotlib.pyplot as plt

linguagens = ["Python","JavaScript","Java","C++","Go"]
popularidade = [95,85,80,70,60]

plt.figure()
plt.bar(linguagens, popularidade)

plt.title("Popularidade de Linguagens")
plt.xlabel("Linguagem")
plt.ylabel("Popularidade")

plt.show()
```

## 📉 5️⃣ Histograma

```python

import matplotlib.pyplot as plt
import random

dados = [random.gauss(50,10) for _ in range(1000)]

plt.figure()
plt.hist(dados, bins=30)

plt.title("Distribuição de Dados")
plt.xlabel("Valores")
plt.ylabel("Frequência")

plt.show()
```

## 🔵 6️⃣ Gráfico de dispersão (Scatter Plot)

```python

import matplotlib.pyplot as plt
import random

x = [random.randint(0,100) for _ in range(50)]
y = [random.randint(0,100) for _ in range(50)]

plt.figure()
plt.scatter(x, y)

plt.title("Gráfico de Dispersão")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

## ✍️ 7️⃣ Manipulação de texto

```python

texto = "Python é uma linguagem muito poderosa"

print(texto.upper())
print(texto.lower())
print("Número de caracteres:", len(texto))
print(texto.replace("poderosa","incrível"))
```

## 🎯 8️⃣ Mini projeto — contador de palavras

```python

frase = input("Digite uma frase: ")

palavras = frase.split()

print("Quantidade de palavras:", len(palavras))
```
